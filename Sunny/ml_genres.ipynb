```python
#Import statements
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

# 1. Data Cleaning

```python
# Load the dataset
df_songs_with_attributes = pd.read_csv("Datasets/songs_with_attributes_and_lyrics.csv")
print(f"Dataset shape: {df_songs_with_attributes.shape}")
df_songs_with_attributes.head()
```

```python
# Load the dataset
df_train = pd.read_csv("Datasets/train.csv")
print(f"Dataset shape: {df_train.shape}")
df_train.head()
```

```python
df_train.info()
```

```python
genres = sorted(df_train['track_genre'].unique())
```

```python
#keep only the columns we need from df_train
df_genres = df_train[["track_id", "track_genre"]]
#rename column
df_genres.rename(columns={'track_id': 'id'}, inplace=True)
df_genres.head()
```

```python
#do the same for df_songs_with_attributes
df_lyrics = df_songs_with_attributes[["id", "artists", "lyrics"]]
df_lyrics.head()
```

```python
# Merging df_lyrics and df_genres

master_df = df_lyrics.merge(df_genres, on='id', how='inner')

master_df.info()
print(master_df.shape)
print(master_df.head())
```

```python
# based on the results, we decided to keep lyrics that have equal or more than 100 words
master_df = master_df[master_df['word_count'] >= 100]

# want to make sure that the lyrics are in english
from langdetect import detect, LangDetectException

def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return 'unknown'

master_df['language'] = master_df['lyrics'].apply(detect_language)
master_df = master_df[master_df['language'] == 'en']
master_df = master_df.drop(columns=['language', 'word_count'])

print(f"Final rows: {len(master_df)}")
```

```python
# Make sure everything looks clean
print(master_df.info())
print(master_df['track_genre'].value_counts())
print(f"Unique genres: {master_df['track_genre'].nunique()}")
```

```python
master_df.head()
```

```python
# Load the dataset
df_parent_genre = pd.read_csv("Datasets/genre_parent_mapping.csv")
print(f"Dataset shape: {df_parent_genre.shape}")
df_parent_genre.head()
```

```python
df_merged = master_df.merge(
    df_parent_genre,
    left_on="track_genre",
    right_on="original_genre",
    how="left"
)
```

```python
#keep only rows that have included = True
print(df_merged["included"].dtype)
print(df_merged["included"].unique())

df_filtered = df_merged[df_merged["included"]].copy()

print(f"Final rows: {len(df_filtered)}")
```
